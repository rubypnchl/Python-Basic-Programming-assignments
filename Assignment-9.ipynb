### 1.	Write a Python program to check if the given number is a Disarium Number?

```python
def disarium_num(n):
    l=[]
    m=n
    while n!=0:
        r=n%10
        l.append(r)
        n=n//10
    l.reverse()
    #print(l)
    s=0
    for i in range(0,len(l)):
        s=s+pow(l[i],i+1)
        #print(s,l[i],i+1)
    #print(s,n)
    if s==m:
        return True
    else:
        return False
n=int(input("enter a number"))
print("is number Disarium",disarium_num(n))
```

### 2.	Write a Python program to print all disarium numbers between 1 to 100?

```python
for i in range(1,100):
    if disarium_num(i):
        print(i)
    
```
